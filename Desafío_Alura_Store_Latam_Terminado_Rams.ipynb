#Primeros pasos del desafío

###Importación de datos



```python
#Importamos la biblioteca de pandas para manipulación de datos

import pandas as pd

#URLs de los archivos CSV correspondientes a las 4 tiendas de Alura Store
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

#Leyendo los archivos CSV y almacenamos cada tienda en DataFrames individuales
tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

###Unificando los DataFrames en uno solo para facilitar comparaciones entre tiendas

```python
#Agregamos una columna que identifique a cada tienda
tienda1['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

#Unificamos todos los DataFrames en uno solo para facilitar el análisis comparativo, usando el método pd.concat
df_tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)
```

```python
#Viendo el tamaño de df_tiendas
df_tiendas.shape
```

```python
#Visualizando los metadatos de los datos de todas las tiendas (df_tiendas)
df_tiendas.info()
```

###Convertimos la columna "Fecha de compra" a formato de fecha

```python
#Conversión usando pd.to_datetime; especificando el formato de fecha
df_tiendas['Fecha de Compra'] = pd.to_datetime(df_tiendas['Fecha de Compra'], format='%d/%m/%Y')
```

```python
#Observando si se realizó el cambio
df_tiendas.info()
```

#1. Análisis de facturación



```python
#Cálculo de ingreso por día en la tienda1
ingreso_por_tienda1 = tienda1.groupby('Fecha de Compra')['Precio'].sum().sort_values(ascending=False)

#Sumatoria de los ingresos para obtener el ingreso total de la tienda
ingreso_total_tienda1 = tienda1['Precio'].sum()

#Imprimir resultado
print(ingreso_total_tienda1)
```

```python
#Cálculo de ingreso total para todas las tiendas

ingresos_por_tienda = df_tiendas.groupby('Tienda')['Precio'].sum().sort_values(ascending=False)

#Imprimir resultados de ingresos por tienda
print("Ingresos por tienda:")
print(ingresos_por_tienda)
```

```python
#Mostrar los ingresos con puntos como separador de miles, para una mejor visualización de las cifras

ingresos_por_tienda_formateados = ingresos_por_tienda.apply(lambda x: f'{x:,.2f}')
print("Ingresos por tienda:")
print(ingresos_por_tienda_formateados)
```

# 2. Ventas por categoría

###Ver elementos dentro de la categoría

```python
#Observamos cuántos elementos existen dentro de la columna de "categoría del producto", ordenado alfabéticamente

categorias_ordenadas = sorted(df_tiendas['Categoría del Producto'].unique())

#Imprimir resultado para observar
print(categorias_ordenadas)
```

```python
#Ver los elementos junto a la cantidad de veces que se vendieron en total (sin distinguir por tiendas)
conteo_categorias = df_tiendas['Categoría del Producto'].value_counts()

#Imprimir resultado para observar
print("Ventas por producto:")
print(conteo_categorias)
```

###Agruparlos para observar la cantidad de ventas por categoría del producto de cada tienda

```python
#Agrupamos por tienda y categoría del producto, y contamos
ventas_por_categoria = df_tiendas.groupby(['Tienda', 'Categoría del Producto'])['Producto'].count().reset_index()

#Renombramos la columna para mejor claridad
ventas_por_categoria.columns = ['Tienda', 'Categoría del Producto', 'Cantidad de Ventas']

#Reorganizamos los datos ordenando de mayor a menor, para visualizarlos mejor
ventas_por_categoria = ventas_por_categoria.sort_values(by=['Tienda', 'Cantidad de Ventas'], ascending=[True, False])

#Mostramos los resultados, separando las categorías por tienda
print("Ventas por producto:")
for tienda in ventas_por_categoria['Tienda'].unique():
    print(f"\n{'='*24} {tienda} {'='*24}") #Dando las separaciones
    print(ventas_por_categoria[ventas_por_categoria['Tienda'] == tienda].reset_index(drop=True)) #Mostrar resultados sin su índice
```

###Sumatoria de cantidad total de ventas por tienda

```python
#Contamos cuántas ventas hay por tienda (una fila = una venta)
cantidad_ventas_por_tienda = df_tiendas['Tienda'].value_counts().sort_index()

#Convertimos a un DataFrame para poder añadir la fila total que deseamos
cantidad_ventas_por_tienda_df = cantidad_ventas_por_tienda.reset_index()
cantidad_ventas_por_tienda_df.columns = ['Tienda', 'Cantidad de Ventas']

#Agregamos una fila final con el total de todas las tiendas
total_general = pd.DataFrame([['Total general', cantidad_ventas_por_tienda.sum()]], columns=cantidad_ventas_por_tienda_df.columns)

#Concatenamos la tabla original con la fila total
ventas_con_total = pd.concat([cantidad_ventas_por_tienda_df, total_general], ignore_index=True)

#Mostramos el resultado
print("Cantidad de ventas totales por tienda:")
print(ventas_con_total)
```

# 3. Calificación promedio (media) de la tienda




```python
#Calculamos la calificación promedio de cada una de las tiendas; redondeado a 2 cifras decimales
calificacion_promedio_tienda = df_tiendas.groupby('Tienda')['Calificación'].mean().round(2)

#Mostramos el resultado
print("Calificación promedio por tienda:")
print(calificacion_promedio_tienda)
```

# 4. Productos más y menos vendidos

```python
#Agrupamos por tienda y producto, y contamos cuántas veces se vendió cada producto
ventas_por_producto = df_tiendas.groupby(['Tienda', 'Producto'])['Producto'].count().reset_index(name='Cantidad de Ventas')

#Ordenamos por tienda y luego por ventas descendente
ventas_por_producto = ventas_por_producto.sort_values(by=['Tienda', 'Cantidad de Ventas'], ascending=[True, False])
```

###Productos más vendidos de las tiendas

```python
#Obtenemos los 5 productos más vendidos por tienda
top5_productos = ventas_por_producto.groupby('Tienda').head(5).reset_index(drop=True)

#Imprimir resultados obtenido
print("Productos más vendidos por tienda:")
for tienda in top5_productos['Tienda'].unique():
    print(f"\n{'='*25} {tienda} {'='*25}") #Dando las separaciones
    print(top5_productos[top5_productos['Tienda'] == tienda].reset_index(drop=True)) #Mostrar resultados sin su índice
```

###Productos menos vendidos de las tiendas

```python
#Ordenamos por tienda y cantidad de ventas, de forma ascendente
menos_vendidos = ventas_por_producto.sort_values(by=['Tienda', 'Cantidad de Ventas'], ascending=[True, True])

#Obtenemos los 5 productos menos vendidos por tienda
bottom5_productos = menos_vendidos.groupby('Tienda').head(5).reset_index(drop=True)

#Imprimir resultados
print("Productos menos vendidos por tienda:")
for tienda in bottom5_productos['Tienda'].unique():
    print(f"\n{'='*25} {tienda} {'='*25}") #Dando las separaciones
    print(bottom5_productos[bottom5_productos['Tienda'] == tienda].reset_index(drop=True)) #Mostrar resultados sin su índice
```

# 5. Envío promedio por tienda

```python
#Agrupamos por tienda y calculamos el promedio del valor de envío
envio_promedio = df_tiendas.groupby('Tienda')['Costo de envío'].mean().sort_values(ascending = False).round(2)

#Mostramos los resultados
print("Costo de envío promedio por tienda:")
print(envio_promedio)
```

#6.Gráficos para comprender mejor los datos analizados

```python
#Importando Matplotlib para graficar
import matplotlib.pyplot as plt

#Importando ticker de Matplotlib.ticker
import matplotlib.ticker as ticker

#Importando Seaborn
import seaborn as sns
```

###Gráfico de barras verticales: ingresos totales por tienda

```python
#Generando el gráfico
fig1, ax1 = plt.subplots(figsize=(10, 5))

#Dándole color al gráfico: fondo y fondo del área de los ejes
fig1.patch.set_facecolor('#2b2b2b')
ax1.set_facecolor('#2b2b2b')

#Dándole sus especificaciones
ingresos_por_tienda.plot(kind='bar', color='#F5B041', ax=ax1)

#Ajustando títulos y sus labels
ax1.set_title("Ingresos totales por tienda", fontsize=16, color = 'white')
ax1.set_ylabel("Ingresos en pesos colombianos", fontsize=14, color = 'white')
ax1.set_xlabel("Tiendas", fontsize=14, color = 'white')

#Ajustando la inclinación de los labels del eje X
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=20, fontsize = 10, color = 'white')

#Ajustando tamaño de las cifras del eje Y
ax1.tick_params(axis='y', labelsize=10, colors = 'white')

#Ajustando el cuadriculado
ax1.grid(axis='y', linestyle='--', alpha=0.5, color = 'white')

#Colocando las cifras encima de cada barra, con separador de miles
ax1.bar_label(ax1.containers[0], fmt='{:,.2f}', color = 'white', fontsize = 10)

#Formatear los valores del eje Y con separador de miles
ax1.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x:,.0f}'))

#Ajustar el rango del eje Y
ax1.set_ylim(0, 1_300_000_000)

#Mostrar gráfico
plt.tight_layout()
plt.show()
```

Este gráfico compara el total de ingresos generados por cada tienda, permitiendo identificar cuáles tienen mejor desempeño financiero.

###Gráfico de barras horizontales: calificaciones promedio por tienda

```python
#Calificaciones promedio por tienda
calificaciones = df_tiendas.groupby('Tienda')['Calificación'].mean().sort_values()

#Crear el gráfico con sus especificaciones
fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.barh(calificaciones.index, calificaciones.values, color='#F5B041')

#Añadiendo los labels
ax.set_title('Calificación promedio por tienda', fontsize=16, color='white')
ax.set_xlabel('Calificación Promedio', fontsize=12, color='white')
ax.set_xlim(3.5, 4.1)

#Fondo gris y texto blanco
fig.patch.set_facecolor('#2E2E2E')
ax.set_facecolor('#2E2E2E')
ax.tick_params(colors='white', labelsize=12)
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')

#Mostrar los valores al final de las barras
for bar in bars:
    width = bar.get_width()
    ax.text(width + 0.01, bar.get_y() + bar.get_height() / 2, f'{width:.2f}', va='center', color='white', fontsize=12)

#Mostrar el gráfico
plt.tight_layout()
plt.show()
```

Se muestran las valoraciones promedio de los clientes por tienda, lo que nos permite evaluar la satisfacción del cliente en cada una de ellas.

###Gráfico de columnas apiladas por categoría: cantidad de ventas por categoría de producto en cada tienda

###Gráfico de líneas: ingresos anuales por tienda



```python
#Paleta de colores que usaremos
colores = ['#FF5733', '#33FF57', '#3357FF', '#FF33A8']

#Extraer los datos
df_tiendas['Año'] = df_tiendas['Fecha de Compra'].dt.year

#Agrupar por tienda y año, sumando ingresos
ingresos_anuales = df_tiendas.groupby(['Tienda', 'Año'])['Precio'].sum().reset_index()

#Crear el gráfico
plt.figure(figsize=(10,5))

#Ajustando el gráfico
for i, tienda in enumerate(ingresos_anuales['Tienda'].unique()):
    datos = ingresos_anuales[ingresos_anuales['Tienda'] == tienda]
    plt.plot(datos['Año'], datos['Precio'] / 1e6, marker='o', label=tienda, color=colores[i]) #División por 1e6 para mostrar en millones

#Cambio en el eje X para que solo se muestren los años y no la mitad de estos
plt.xticks(sorted(ingresos_anuales['Año'].unique()))

#Título del gráfico con sus características
plt.title('Ingresos anuales por tienda', fontsize=16, color='white')

#Ajustando los labels
plt.xlabel('Años', fontsize=14, color='white')
plt.ylabel('Ingresos en millones de pesos colombianos', fontsize=14, color='white')

#Cambiar color de los números de los ejes X/Y
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')

#Ajustando el fondo para el gráfico y figura
plt.gcf().set_facecolor('#333333')
plt.gca().set_facecolor('#444444')

#Añadiendo una leyenda para mejor interpretación
plt.legend(title='Tienda')

#Ajustando el cuadriculado
plt.grid(True, linestyle='--', alpha=0.5)

#Mostrar el gráfico
plt.tight_layout()
plt.show()
```

Se visualiza la evolución de los ingresos anuales de cada tienda entre 2020 y 2023, mostrando tendencias de crecimiento o caída.

#Extra: análisis del desempeño geográfico

###Gráfico de dispersión: distribución geográfica de las ventas por tienda

```python
#Importando Line2D de la biblioteca MatPlotLib.lines
from matplotlib.lines import Line2D

#Paleta de colores por tienda
colores = {
    'Tienda 1': '#FF5733',
    'Tienda 2': '#33FF57',
    'Tienda 3': '#3357FF',
    'Tienda 4': '#FF33A8'
}

#Crear un gráfico de dispersión
plt.figure(figsize=(10, 5))

#Graficamos cada tienda por separado
for tienda in df_tiendas['Tienda'].unique():
    datos = df_tiendas[df_tiendas['Tienda'] == tienda]

    plt.scatter(
        datos['lon'], datos['lat'],
        alpha=0.6,  # Transparencia mejorada (antes 0.4)
        s=datos['Precio'] / 1500,  # Círculos más pequeños (antes /500)
        color=colores[tienda],
        label=tienda
    )

#Ajustando el título del gráfico
plt.title('Distribución geográfica de las ventas por tienda', fontsize=16, color = 'white')

#Añadiendo los labels
plt.xlabel('Longitud', fontsize = 14, color = 'white')
plt.ylabel('Latitud', fontsize = 14, color = 'white')

#Ajustando la leyenda personalizada con círculos del mismo color
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label=tienda,
           markerfacecolor=color, markersize=8, alpha=0.6)  # Circulo más pequeño y alpha coherente
    for tienda, color in colores.items()
]
plt.legend(handles=legend_elements, title='Tienda', loc='upper right')

#Ajustando el cuadriculado del gráfico
plt.grid(True, linestyle='--', alpha=0.5)

#Fondo del gráfico y de la figura, con colores que combinen con la paleta
plt.gca().set_facecolor('#444444')
plt.gcf().set_facecolor('#333333')

#Cambiar color de los números de los ejes X/Y
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')

#Mostrar el gráfico
plt.tight_layout()
plt.show()
```

Muestra en un mapa la ubicación de cada tienda (con su latitud/longitud), con círculos que representan el volumen de ingresos por ubicación.

###**Informe final: análisis de desempeño de tiendas**

####Introducción
El presente informe tiene como objetivo analizar el desempeño de las cuatro tiendas vistas anteriormente, con el fin de ayudar al Sr. Juan a tomar una decisión estratégica: vender una de las tiendas para poder emprender un nuevo negocio.

Para ello, hemos analizado diversos factores clave, como ingresos totales, productos más vendidos, calificaciones promedio de los clientes, categorías más populares y coste promedio de envío; todo esto apoyándose en visualizaciones gráficas que permiten una correcta comparación entre las tiendas.

####Desarrollo

#####Ingresos totales por tienda:
El análisis de ingresos revela que la Tienda 3 genera la mayor facturación entre todas las tiendas, mientras que la Tienda 4 tiene los ingresos más bajos. Esto se refleja en el gráfico de barras verticales separadas, donde la diferencia entre ambas tiendas es considerable. La Tienda 1 y la Tienda 2 presentan ingresos intermedios, siendo relativamente similares.

#####Calificaciones promedio de clientes:
El análisis de las reseñas de clientes muestra que la Tienda 2 obtiene la mejor calificación promedio, lo que sugiere una buena experiencia de compra. Por el contrario, la Tienda 4 tiene las peores valoraciones, lo cual representa una debilidad significativa, especialmente si consideramos que también tiene el menor volumen de ingresos.

#####Categorías de productos más vendidas:
El gráfico de barras verticales agrupadas indica que la Tienda 3 lidera en la venta de varias categorías, especialmente en las más rentables como “Electrónica” y “Accesorios”. En contraste, la Tienda 4 muestra ventas bajas y dispersas en las categorías, sin destacar en ninguna en particular.

#####Productos más y menos vendidos:
El análisis de productos individuales muestra que la Tienda 3 y la Tienda 1 concentran la mayor cantidad de unidades vendidas de los productos más populares. Nuevamente, la Tienda 4 aparece con los menores volúmenes de venta, incluyendo varios productos con ventas nulas o muy bajas.

#####Coste de envío promedio:
En cuanto al coste de envío, la Tienda 4 tiene el promedio más alto, lo que puede afectar negativamente la satisfacción del cliente y su disposición a comprar. En cambio, la Tienda 2 tiene el envío promedio más bajo, lo cual contribuye a su alta calificación por parte de los clientes.

####Conclusiones:
Con base en los datos analizados, consideramos que la tienda que el Sr. Juan debería vender es la tienda 4. Esta recomendación que hicimos se fundamenta en los siguientes 5 puntos clave:

1. Menor ingreso total, lo que implica un bajo rendimiento económico.

2. Peores calificaciones de clientes, lo que refleja una experiencia insatisfactoria.

3. Baja competitividad en todas las categorías de productos, sin ninguna especialización clara.

4. Mayores costes de envío promedio, lo cual representa una desventaja operativa.

5. Baja rotación de productos, incluyendo muchos artículos con pocas o ninguna venta.

Además de eso, podemos observar que la tienda 3 demuestra ser la más sólida en términos de ingresos, variedad de productos vendidos y penetración en categorías clave. La tienda 2, si bien no es la que más vende, muestra una muy buena relación con los clientes y eficiencia en costos, mientras que la tienda 1 mantiene un buen equilibrio entre ventas y desempeño general.

Por lo tanto, consideramos que vender la tienda 4 permitirá al Sr. Juan liberarse de la tienda con peores indicadores, y enfocarse en aquellas mencionadas que tienen mayor potencial de crecimiento o estabilidad operativa.
